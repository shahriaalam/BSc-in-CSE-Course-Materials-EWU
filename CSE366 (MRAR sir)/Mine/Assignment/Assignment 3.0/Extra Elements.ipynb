```python
import pandas as pd
import numpy as np
from tabulate import tabulate

# Path to your results.csv file
results_csv_path = '/content/drive/MyDrive/Assignment 3/runs/detect/train/results.csv'

# Load the CSV file
results_df = pd.read_csv(results_csv_path)



# Strip any leading/trailing whitespace from column names
results_df.columns = results_df.columns.str.strip()


# Extract the last epoch's results
last_epoch = results_df.iloc[-1]

# Define the correct column names
precision_col = 'metrics/precision(B)'
recall_col = 'metrics/recall(B)'
map50_col = 'metrics/mAP50(B)'

if precision_col in results_df.columns and recall_col in results_df.columns:
    # Calculate F1 score using numpy
    precision = last_epoch[precision_col]
    recall = last_epoch[recall_col]
    f1_score = 2 * (precision * recall) / (precision + recall)

    # Compile metrics into a dictionary
    metrics = {
        'Model': ['YOLOv8'],
        'Accuracy': [f"{last_epoch[map50_col]:.2%}"],  # Assuming mAP50 as accuracy
        'Precision': [f"{precision:.2%}"],
        'Recall': [f"{recall:.2%}"],
        'F1 score': [f"{f1_score:.2%}"]
    }

    # Create a DataFrame for the table
    metrics_df = pd.DataFrame(metrics)

    # Print the table using tabulate for better formatting
    print(tabulate(metrics_df, headers='keys', tablefmt='grid'))
else:
    print(f"Column names '{precision_col}' or '{recall_col}' not found in the DataFrame")
```
